```python
import matplotlib.pyplot as plt
import numpy as np
s=int(input("size"))
n=int(input("iterations"))
points=[(0,0),(s,0),(s/2,0.866*s)]
x=[]
y=[]
counter=0
while counter<=n:
  np.random.shuffle(points)
#print(points)
  dummy=points[:2]
#print(dummy)
  xind=yind=0
  for i in dummy:
    xind+=i[0]
  for j in dummy:
    yind+=j[1]
  NewPoint=(xind/2,yind/2)
  points.append(NewPoint)
  print(points)
  counter+=1

for i in points:
    a=i[0]
    x.append(a)
for i in points:
    a=i[1]
    y.append(a)
xpoints = np.array(x)
ypoints = np.array(y)

plt.plot(xpoints, ypoints, 'o')
plt.show()
```
